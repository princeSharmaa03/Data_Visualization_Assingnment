5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate synthetic dataset with correlated features
np.random.seed(42)

# Generate random data
x = np.random.rand(100)
y = 2.5 * x + np.random.randn(100) * 0.5  # Correlated with x
z = 0.5 * x + 0.2 * y + np.random.randn(100) * 0.5  # Correlated with both x and y

# Combine features into a DataFrame
import pandas as pd
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', center=0)

# Title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, output_file, show
import numpy as np
import pandas as pd

# Set up the output file (the plot will be saved as an HTML file)
output_file("scatter_plot.html")

# Generate random data for x, y, sizes, and colors
np.random.seed(42)  # For reproducibility
n = 100  # Number of points

x = np.random.random(n) * 10  # Random x values between 0 and 10
y = np.random.random(n) * 10  # Random y values between 0 and 10
sizes = np.random.random(n) * 100  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors from a list

# Create a DataFrame for better visualization (optional)
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add a scatter plot to the figure with different sizes and colors for markers
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=df, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate random data for the histogram (1000 values from a normal distribution)
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Set up the output file (the plot will be saved as an HTML file)
output_file("histogram.html")

# Create a Bokeh figure
p = figure(title="Histogram of Normally Distributed Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           tools="pan,box_zoom,reset,save")

# Add the histogram to the plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import numpy as np
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.io import show
from bokeh.palettes import Viridis256
from bokeh.layouts import column

# Generate the data
data_heatmap = np.random.rand(10, 10)  # Random 10x10 matrix
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Set up the output file (the plot will be saved as an HTML file)
output_file("heatmap.html")

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', 
           x_range=(0, 1), y_range=(0, 1), tools="hover,pan,box_zoom,reset")

# Define a linear color mapper
from bokeh.models import LinearColorMapper
from bokeh.palettes import Viridis256

color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap using image()
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper, width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
